```python
from paddleocr import PaddleOCR

ocr = PaddleOCR(use_angle_cls=True, lang='en')

image_path = "number_plate.jpg"

result = ocr.ocr(image_path, cls=True)

for line in result:
    for word_info in line:
        text = word_info[1][0]
        print(text)
```

```python
import cv2
from yolov5 import YOLOv5
import pytesseract
import torch

model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
cap = cv2.VideoCapture('demo1.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)  
    
    if len(results.xyxy[0]) > 0:
        for *box, conf, cls in results.xyxy[0]:
            xmin, ymin, xmax, ymax = map(int, box)

            plate_image = frame[ymin:ymax, xmin:xmax]
            
            text = pytesseract.image_to_string(plate_image, config='--psm 8')
            print("Detected License Plate:", text)

            cv2.imwrite('detected_license_plate.jpg', plate_image)
            cv2.imshow('License Plate', plate_image)

            break

    cv2.imshow('Video', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
